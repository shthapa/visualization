```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Using Scatter diagram to view the plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
# Using Bar diagram to view the plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
# using barh diagram to view the plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y)
```

```python
#Using histogram graph to view the chart
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
# adding title and lable names
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
# adding color combination to the chart
font1 = {'family': 'sans-serif','color':'blue','size':20}
font2 = {'family': 'monospace', 'color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# adding color combination with font size
font1 = {'family': 'sans-serif','color':'blue','size':20}
font2 = {'family': 'monospace', 'color':'green','size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc='left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# libraries
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# initiate figure
fig, ax = plt.subplots(figsize=(10, 8), dpi=120)

def update(frame):
    ax.clear()
    ax.scatter(
      1+frame, 10+frame*10,
      s=600, alpha=0.5,
      edgecolors="black"
    )
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 100)
    return fig, ax

ani = FuncAnimation(fig, update, frames=range(10))
ani.save("my_animation.gif", fps=5)
```

```python
# Create time series data
dates = pd.date_range(start="2025-01-01", periods=100, freq='D')
values = np.cumsum(np.random.randn(100))  # cumulative sum to simulate trend

# Plot
fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(dates, values)

# Customize
ax.set_title("Sample Time Series")
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
